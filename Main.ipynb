```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel('social_robot4learning.xlsx', sheet_name = 'Sheet1')
data.head()
```

```python
plt.figure(figsize=(20,10))
ax = sns.countplot(x="wtype", data=data, order=['2004-2009','2010', '2011', '2012', '2013', '2014', 
                                              '2015', '2016', '2017', '2018', '2019', '2020'], palette="plasma")
```

## Papers per country

```python
data.Country.unique()
```

```python
df_plot = data.groupby(['wtype', 'Country']).size().reset_index().pivot(columns='Country', index='wtype', values=0)
df_plot.plot(kind='bar', stacked=True, cmap='tab20b')
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```

## Robot Used

```python
data['Robot/s used (?)'].unique()
```

```python
df_plot = data.groupby(['Robot/s used (?)', 'wtype']).size().reset_index().pivot(columns='Robot/s used (?)', index='wtype', values=0)
df_plot.plot(kind='bar', stacked=True, cmap='tab20b')
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```

# Nb of students per robots *

*Number of students concurrently interacting with the robot

```python
df_plot = data.groupby(['students_nb', 
                             'wtype']).size().reset_index().pivot(columns='students_nb', index='wtype', values=0)

df_plot.plot(kind='bar', stacked=True, cmap=plt.cm.get_cmap('plasma', 4))
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```

# Communicative gestures vs functional actions *

*Communicative gestures refer to expressive motions, as the opposite of motions that are associated to manipulation of an object of displacement in space, here coded as functional actions. We also coded specifically Diectic gestures as D.

```python
df_plot = data.groupby(['Communicative gestures vs functional actions', 
                             'wtype']).size().reset_index().pivot(columns='Communicative gestures vs functional actions', index='wtype', values=0)

df_plot.plot(kind='bar', stacked=True, cmap=plt.cm.get_cmap('plasma', 4))
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```

# Retention *

'Yes' if the study contains a retention test.

# Material *
Describe which material is used in additon to the robot
We proposed to code for :
- PC
- Tangibles (cubes, balance beam etc.)
- Screen
- Tablet

some papers combine the material

```python
df_plot = data.groupby(['Matos', 
                             'wtype']).size().reset_index().pivot(columns='Matos', index='wtype', values=0)
df_plot[['-', 'tangible(s)', 'tangible(s)+screen', 'tangible(s)+tablet', 'PC', 'PC+tangible(s)', 'screen', 'screen+tablet', 'tablet']].plot(kind='bar', stacked=True, cmap=plt.cm.get_cmap('RdBu', 9))
plt.legend(loc='center left', bbox_to_anchor=(1.0, 0.5))
plt.show()
```
